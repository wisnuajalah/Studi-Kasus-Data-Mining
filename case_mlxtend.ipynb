#### **Tranformasi datanya**

```python
te = TransactionEncoder()
te_try = te.fit(dataset).transform(dataset)
```

#### **Kerangka data**

```python
df = pd.DataFrame(te_try, columns=te.columns_)
df
```

#### **Menggunakan algoritma Apriori**
kami sudah menghitung secara manual untuk mendapatkan min_support dari data sample (30 data)

```python
frequent_itemsets = apriori(df, min_support=0.06, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

##### **Membentuk sebuah relasi**

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 2) & (frequent_itemsets['support'] >= 0.06) ]
```
